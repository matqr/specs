# Steps to add new samples
1. Run entire pipeline and save images in `v1/`
2. Manually go through images and write down the name (`uuid`) of the images that are not suitable.
3. Remove this images and save the resulting ones in `v2/`
4. Launch this notebook to download missing samples. Repeat as needed

```python
df_city = df_clustered[df_clustered["city_id"] == 1152554349]

# count the number of rows in each cluster that were removed
cluster_counts = df_removed["cluster_label"].value_counts()

# keep track of the samples
df_sampled_city = pd.DataFrame()

# sample the missing amount from each cluster
for cluster_label, count in cluster_counts.items():
    print(f"Sampling {count} samples from cluster {cluster_label}")

    # the available samples are the rows within this cluster that have not been already downloaded (metadata_downloaded)
    available_samples = df_city[
        (df_city["cluster_label"] == cluster_label)
        & (~df_city["uuid_x"].isin(df_metadata_downloaded["uuid"]))
    ]

    sampled_from_cluster = available_samples.sample(n=count, replace=False)

    df_sampled_city = pd.concat([df_sampled_city, sampled_from_cluster])

df_sampled_city
```

```python
# final selected images 100 images
def get_file_names_without_extension(folder_path):
    file_names = []
    for file_name in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, file_name)):
            name, _ = os.path.splitext(file_name)
            file_names.append(name)
    return file_names

# Example usage
folder_path = "data/svi/sampled/Final_100/Santiago"
final_100 = get_file_names_without_extension(folder_path)

folder_path = "data/svi/sampled/Final_80/Santiago"
final_80 = get_file_names_without_extension(folder_path)
```

```python
# add lists
to_remove = final_100 + final_80 + REMOVED_SVIS_1 + REMOVED_SVIS_2 + REMOVED_SVIS_3 + REMOVED_SVIS_4
```

```python
# remove already sampled images as well as the ones to remove
df_removed = df_city[~df_city["uuid_x"].isin(to_remove)]
df_removed.shape
```

```python
sampled_from_cluster = df_removed.sample(n=200, replace=False)
sampled_from_cluster.shape
```
